```python
import os
os.environ["MLFLOW_ALLOW_FILE_STORE"] = "true"

import pandas as pd
import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
mlflow.set_tracking_uri("file:./mlruns")
mlflow.set_experiment("iris_classification")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
train_df = pd.read_csv("iris_train_preprocessed.csv")
test_df = pd.read_csv("iris_test_preprocessed.csv")

train_df.head()
test_df.head()
```

```python
X_train = train_df.drop("target", axis=1)
y_train = train_df["target"]
X_test = test_df.drop("target", axis=1)
y_test = test_df["target"]
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average="weighted")
rec = recall_score(y_test, y_pred, average="weighted")
f1 = f1_score(y_test, y_pred, average="weighted")

print("Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1-score:", f1)
```

```python
with mlflow.start_run():
    mlflow.log_param("model_name", "RandomForestClassifier")
    mlflow.log_param("random_state", 42)
    mlflow.log_metric("accuracy", acc)
    mlflow.log_metric("precision", prec)
    mlflow.log_metric("recall", rec)
    mlflow.log_metric("f1_score", f1)
    mlflow.sklearn.log_model(model, "model")
```

```python
from mlflow.tracking import MlflowClient
client = MlflowClient()
print(client.search_experiments())
print(client.search_runs(experiment_ids=["1"]))
```
